# Machine Learning

Machine learning is basically training a model, known as 'model' to make predictions or generate content using data. We feed it data, and it returns a prediction, it can be categorised as supervised, unsupervised, reinforcement, or generative AI.

## Supervised Learning
Labelled data to make predictions, often for regression (predicting numerical values) or classification (categorising data)

## Unsupervised Learning
Identiffies patterns in unlabeled data, commonly using clustering to group similar data points.

## Generative AI
Creates new content such as text, images, or music, by learning patterns from existing data nad mimicking them.

To summarise:

Imagine teaching a computer to predict house prices, this is how it would work.
1. The data:
    Feature ($X$) are like clues or characteristics we use to make predictions, for a house this might be
       - size of the house
       - number of bedrooms
       - location
       - age of the house
    Target ($y$) are what we're trying to predict (house price)
2. Learning process:
    Giving the computer a big spreadsheet with lots of houses where we know the features and the selling prices.
    The computer looks for pattern in this data, like 'bigger housing costing more' or 'location affecting house prices'
4. Training vs Testing:
    Think of this studying for a test where;
       - training data - practice problems you use to learn
       - testing data - exam with new questions
    Keep these separate because we want to make sure our model can make good predictions on houses it hasn't seen before, just like a real test/exam.
6. Making prediction:
    After training, when we show the computer new houses with the features, it uses what it learned to make a prediction about the price.

# Regression
Predicting a number based on other information, e.g. predicing house price (a num) based on size, location, etc...

## Linear Regression
Simplest type of prediction model, thinknof it as finding the 'line of best fit' through your data points, it assumes there's a straight-line relationship between input and output.
